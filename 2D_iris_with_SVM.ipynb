```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from pylab import *

from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.data
```

```python
x= iris.data[:, :2]
y= iris.target
```

```python
y
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=y)
```

```python
clf=SVC()
clf.fit(x, y) #why input y? y is the answer
```

```python
xx, yy = np.meshgrid(np.arange(3,8.5,0.02),np.arange(1.5,5,0.02))
z=clf.predict(np.c_[xx.ravel(), yy.ravel()])
z=z.reshape(xx.shape)
plt.contourf(xx, yy, z, cmap=plt.cm.coolwarm, alpha=0.2)
plt.scatter(x[:,0], x[:,1], s=50, c=clf.predict(x))
```
