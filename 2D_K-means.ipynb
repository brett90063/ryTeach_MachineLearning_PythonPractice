```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from pylab import *

x = np.random.rand(100,2)

from sklearn.cluster import KMeans
clf = KMeans(n_clusters=10)
clf.fit(x)
clf.labels_
scatter(x[:,0], x[:,1], s=100, c=clf.labels_)
show()
```

```python
xx, yy = np.meshgrid(np.arange(0,1.2,0.02),np.arange(0,1.2,0.02))
z=clf.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
z=z.reshape(xx.shape)   #shape:查看矩陣或者數組的維數
                        #reshape:給予數組一個新的形狀，而不改變它的數據
plt.contourf(xx, yy, z, cmap=plt.cm.hot, alpha=0.5)  #cmap color: https://matplotlib.org/examples/color/colormaps_reference.html
plt.scatter(x[:,0], x[:,1], s=50, c=clf.predict(x))
```
