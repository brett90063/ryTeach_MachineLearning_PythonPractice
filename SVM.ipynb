```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from pylab import *
from sklearn.datasets import make_classification

x, y = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, n_classes=3)  #
plt.scatter(x[:,0], x[:,1], s=50, c=y)
```

```python
clf=SVC()
clf.fit(x, y)
plt.scatter(x[:,0], x[:,1], s=50, c=clf.predict(x))
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=clf.predict(x)-y)
```

```python
xx, yy = np.meshgrid(np.arange(-4,4,0.02),np.arange(-3,4,0.02))
z=clf.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
z=z.reshape(xx.shape)   #shape:
                        #reshape:
plt.contourf(xx, yy, z, cmap=plt.cm.coolwarm, alpha=0.2)
plt.scatter(x[:,0], x[:,1], s=50, c=clf.predict(x))
```
